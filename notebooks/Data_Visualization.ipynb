```python
data_dir='../Data/PlantDiseasesDataset/'
```

```python
import os
import random
import matplotlib.pyplot as plt
os.listdir(data_dir)
```

```python
def get_different_images_dir(data_dir):
    classes=[]
    for sub_dir in os.listdir(data_dir):
        classes.append(sub_dir)
    sub_dir_data_train=[]
    for sub_dir in classes:
        sub_dir_data_train.append(data_dir+sub_dir+'/train/')
    sub_sub_dir = []
    labels = []
    for i in range(0, len(sub_dir_data_train)-1):
        for j in os.listdir(sub_dir_data_train[i]):
            sub_sub_dir.append(sub_dir_data_train[i]+j)
            labels.append(j)
    images_path=[]
    for i, dir in enumerate(sub_sub_dir):
        images_path.append(dir+'/'+random.choice(os.listdir(sub_sub_dir[i])))
    return images_path, labels

def visualize_images(img_path_list):
    plt.figure(figsize=(20, 7))
    for i in range(0, len(img_path_list[0])-1):
        #img_array = plt.imread(img_path)
        plt.subplot(3, 9, i+1)
        plt.imshow(plt.imread(img_path_list[0][i]))
        plt.axis(False)
        plt.title(img_path_list[1][i])
        plt.tight_layout()
    plt.show()
```

```python
images_path_list = get_different_images_dir(data_dir)
visualize_images(images_path_list)
```

```python
import os
import shutil
import random

# Step 1: Create a new directory
new_dir = ['C:/Users/User/Documents/Machine/Plants-Classification-Project/src/Crop_Disease_Detection_Model/Data/PlantDiseasesDataset_10_percent/Apple/valid/' + k for k in os.listdir('C:/Users/User/Documents/Machine/Plants-Classification-Project/src/Crop_Disease_Detection_Model/Data/PlantDiseasesDataset/Apple/valid')]
print(new_dir)

for i in range(0, 4):
    
    if not os.path.exists(new_dir[i]):
        os.makedirs(new_dir[i])
    
    # Step 2: Get a list of all image files in the original directory
    image_files = [f for f in os.listdir(original_dir[i]) if f.endswith('.JPG') or f.endswith('.PNG') or f.endswith('.jpg') or f.endswith('.png')]
    random.shuffle(image_files)
    
   # Step 3: Calculate 10% of the total number of images
    num_images = len(image_files)
    print(f'NUM_IMAGES:{num_images}')
    num_images_to_copy = int(num_images * 0.05)
    print(f'COPYING {num_images_to_copy} INTO NEW DIRECTORY')

    # Step 4: Copy the first 10% of images to the new directory
    for j in range(num_images_to_copy):
        src_path = os.path.join(original_dir[i], image_files[j])
        dst_path = os.path.join(new_dir[i], image_files[j])
        shutil.copy(src_path, dst_path)
```
